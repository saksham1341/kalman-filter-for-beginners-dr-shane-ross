```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def average_filter(inp):
    prev = 0

    for k, i in enumerate(inp):
        alpha = k / (k + 1)
        res = alpha * prev + (1 - alpha) * i

        yield res

        prev = res
```

```python
def main():
    t = 100
    base_voltage = 14.4
    noise_mag = 4

    noise = (np.random.rand(t) - 0.5) * noise_mag
    input_signal = noise + base_voltage

    average_filter_output = [x for x in average_filter(input_signal)]

    plt.title("Average Filter")
    plt.xlabel("Time")
    plt.ylabel("Volate")
    plt.plot(input_signal, label = "Input Signal", color = "b", marker = ".")
    plt.plot(average_filter_output, label = "Average Filter Value", color = "r", marker = ".")
    plt.legend()
    plt.show()
    plt.close()

    print(f"Final Average Filter values is {average_filter_output[-1]}")
```

```python
if __name__ == "__main__":
    main()
```
