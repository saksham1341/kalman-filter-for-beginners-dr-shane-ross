```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def generate_input(start, noise):
    prev = start
    
    for x in noise:
        curr = prev + x
        yield curr
        prev = curr
```

```python
def low_pass_estimate(inp, alpha):
    prev = inp[0]

    for x in inp:
        curr = alpha * prev + (1 - alpha) * x
        yield curr
        prev = curr
```

```python
def main():
    t = 150
    base_signal = 14
    noise_mag = 4
    noise = (np.random.random(t) - 0.5) * noise_mag
    alpha = [0.1, 0.3, 0.5, 0.75, 0.99]

    inp = [x for x in generate_input(base_signal, noise)]
    outs = [[x for x in low_pass_estimate(inp, a)] for a in alpha]

    plt.title("Low Pass Estimate")
    plt.plot(inp, label = "Input Signal")
    for ind, o in enumerate(outs):
        plt.plot(o, label = f"Alpha = {alpha[ind]}")
    plt.legend()
    plt.show()
    plt.close()
```

```python
if __name__ == "__main__":
    main()
```
