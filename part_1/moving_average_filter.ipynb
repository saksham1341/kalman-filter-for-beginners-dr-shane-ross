```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def generate_input(start, noise):
    prev = start
    
    for x in noise:
        curr = prev + x
        yield curr
        prev = curr
```

```python
def moving_average_filter(arr, n):
    prev = sum(arr[:n]) / n
    yield prev
    
    low = 0
    for high in range(n, len(arr)):
        curr = prev + (arr[high] - arr[low]) / n
        yield curr
        prev = curr
        low += 1
```

```python
def main():
    t = 150
    n = 4
    noise_mag = 4
    noise = (np.random.random(t) - 0.5) * noise_mag
    base_signal = 14
    
    inp = [x for x in generate_input(base_signal, noise)]
    out = [x for x in moving_average_filter(inp, n)]

    plt.title(f"Moving Average (n = {n})")
    plt.plot(range(0, t), inp, label = "Input Signal")
    plt.plot(range(n - 1, t), out, label = "Moving Average")
    plt.legend()
    plt.show()
    plt.close()
```

```python
if __name__ == "__main__":
    main()
```
