```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def generate_data(n, low = 0, high = 100):
    return low + np.random.random((1, n)).flatten() * high
```

```python
def moving_average_filter(arr, n):
    prev = arr[:n].sum() / n
    yield prev
    
    low = 0
    for high in range(n, len(arr)):
        curr = prev + (arr[high] - arr[low]) / n
        yield curr
        prev = curr
        low += 1
```

```python
def main():
    t = 100
    n = 3
    inp = generate_data(t)
    out = [x for x in moving_average_filter(inp, n)]

    plt.title(f"Moving Average (n = {n})")
    plt.plot(range(0, t), inp, label = "Input Signal")
    plt.plot(range(n - 1, t), out, label = "Moving Average")
    plt.legend()
    plt.show()
    plt.close()
```

```python
if __name__ == "__main__":
    main()
```
